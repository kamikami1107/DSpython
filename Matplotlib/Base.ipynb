````python
# データセットを再ダウンロード

df = pd.read_csv('/work/Pandas/movie_dataset/tmdb_5000_movies.csv')
````

````python
df.head()
````

````python
# 西暦を抜き出すには？

sample = df['release_date'][0]
sample[:4]
````

````python
# 関数を作成
# release_dateから西暦を抜き出す関数

def get_year(release_date):
    '''
    Parameters
    ----------------
    release_date:
    映画が公開された年月日
    
    Return
    ----------------
    year:
    映画が公開された西暦
    '''
    
    return release_date[:4]
    
````

````python
# 確認

get_year(df['release_date'][0])
````

````python
# lambda関数で作成する

date = lambda x: x[:4]
````

````python
# 確認

date(df['release_date'][0])
````

````python
# apply関数に入れてdfに適用
# 適用したものをdf['year']としてdfに新規作成

df['year'] = df['release_date'].apply(lambda x: x[:4])
df.head()
````

2. revenueの年推移

````python
# revenueがnanのものと0のものは除く
# revenueがnanのもの、または、0のもの、を除いたもの

df[~((df['revenue'].isna()) | (df['revenue'] == 0))]
````

````python
# revenueがnanのものと0のものは除く、その２
# revenueがnanではないもの、かつ、0ではないもの

df[(df['revenue'] != np.nan) & (df['revenue'] != 0)]
````

````python
# 上記の条件を当てはめてdfを更新

df = df[(df['revenue'] != np.nan) & (df['revenue'] != 0)]
df.head()
````

````python
# 西暦ごとの各columnsの中央値(median)を算出し、
# その中のrevenueを表示

revenue_median = df.groupby('year').median()['revenue']
revenue_median
````

````python
# 平均

revenue_mean = df.groupby('year').mean()['revenue']
revenue_mean
````

````python
# 最大値

revenue_max = df.groupby('year')['revenue'].max()
revenue_max
````

#### Memo

今回の場合のmaxのコード内の```['revenue']```の位置がmeanやmedianと違うのは、

meanやmedianは、各columnsを西暦毎に計算した結果で、最後にその中でrevenueというcolumnsを選んできている。

```df.groupby('year').median()['revenue']```


に対して、maxはまず**どのcolumnsの最大値なのか**ということを指定しなければならない。
その為```groupby('year')```で年代毎に分類したら、その中の**revenueに対して**の最大値という指定の方法をしてあげる必要がある。

その為にコードが、

```df.grouby('year')['revenue'].max()```

となる。

````python
# グラフ描写

plt.plot(revenue_median, label='median')
plt.plot(revenue_mean, label='mean')
plt.plot(revenue_max, label='max')
plt.legend()
plt.title('Trend of revenue')
````

### plt.subplots()

一番よく使う表示方法

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, x**2)
axes[1].plot(x, x**3)
````

````python
# 4つのグラフを同時に描写

fig, axes = plt.subplots(2, 2)# 2*2の二次元配列
axes[0, 0].plot(x, x)
axes[0, 1].plot(x, x**2, 'r')
axes[1, 0].plot(x, x**3, 'y')
axes[1, 1].plot(x, x**4, 'g')
fig.tight_layout()
````

````python
# 一つだけのグラフ描写にも使える

fig, axes = plt.subplots()# defaultは(1, 1)
axes.plot(x, x**2)
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, x**2, label='x2')
axes[1].plot(x, x**3, label = 'x3')
axes[0].set_xlabel('xlabel1')
axes[0].set_ylabel('ylabel1')
axes[1].set_xlabel('xlabel2')
axes[1].set_ylabel('ylabel2')
axes[0].set_title('X2')
axes[1].set_title('X3')
# axes[0].set_xticks()
# axes[1].set_xticks()
axes[0].legend()
axes[1].legend()
fig.tight_layout()
````
