```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_dataset/tmdb_5000_movies.csv')
df.head()
```

### .groupby('columns').[統計量]

指定したcolumnsの統計量を計算してくれる

```python
# original_languageのグループごとの平均値を計算する

df.groupby('original_language').mean()
```

```python
# 日本語の物だけを抜き出す

df.groupby('original_language').mean().loc['ja']
```

```python
# 検算
# original_languageが日本語のものだけのbudgetの平均値を求めてみる

a = df[df['original_language'] == 'ja']['budget'].mean()

# 小数点以下の桁数指定
print('{:.3f}'.format(a))
# 指数表記で小数点以下の桁数指定
print('{:6e}'.format(14293606.188))
```

```python
# しかしながら、上記の平均値はbudgetが0のものが含まれてしまっている
# budgetが0のものを除いて、平均値を算出

df1 = df[df['budget'] != 0].groupby('original_language').mean()
df1
```

```python
# この書き方でもOK
# budgetが0のものを除いて、平均値を算出

df2 = df[~(df['budget'] == 0)].groupby('original_language').mean()
df2
```

```python
# データ数で検算

len(df1) == len(df2)
```

```python
# groupがjaのものの平均値で検算

df1.iloc[:]['budget'].loc['ja'] == df2.iloc[:]['budget'].loc['ja']
```

```python
# budgetが0のものの統計量を、original_languageグループ内でカウント

df[df['budget'] == 0].groupby('original_language').count()
```

```python
# budgetが0のものの統計量を、original_languageグループ内でカウント
# その中からbudgetだけを抜き出す

df[df['budget'] == 0].groupby('original_language').count()['budget']
```

```python
# origial_languageグループ内で、各columnsの統計量を見る
# 平均値だけじゃなくて全部

df.groupby('original_language').describe()
```

```python
# origial_languageグループ内で、各columnsの統計量を見る
# 今回は更にcolumnsをrevenueで指定

df.groupby('original_language').describe()['revenue']
```

```python
# 上記のグループでrevenue(評価)が最も高いものを抜粋

df.groupby('original_language')['revenue'].max()
```

```python
# 更に各言語で評価が高かったのはどの映画なのかを出す
# 最高評価を出しているrevenueに紐づいた映画のインデックスを取ってくる
# maxの前にidxを付けてidxmaxとすると最大値のインデックスを取れる

df.groupby('original_language')['revenue'].idxmax()
```

```python
# 更にインデックスをilocで取ることにより、インデックスのSeriesを取ってこれる
# 頻出の形

df.iloc[df.groupby('original_language')['revenue'].idxmax()]
```

```python
max_rev_list = df.iloc[df.groupby('original_language')['revenue'].idxmax()]
```

```python
# 上記の中で日本語の作品を取ってくる

max_rev_list[max_rev_list['original_language'] == 'ja']
```
