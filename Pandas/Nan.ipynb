### .dropna()

対象のdataframeのnanを除外する

```python
df.dropna()
```

```python
# nanを探してみる

df.describe()
```

```python
print(len(df))
```

```python
# 全データ数4803に対して、runtimeのcountが4801の為、
# 欠損値（nan）を含んでいる可能性が高い
# runtimeがnanになっているデータを取り出す

df[np.isnan(df['runtime'])]# np.isnan()で使えるのは数値項目のみ
```

```python
# runtimeがnanになっている以外のデータを取り出す

rt1 = df[~(np.isnan(df['runtime']))]
rt1
```

```python
# runtimeがnanになっている以外のデータを取り出す、その２
# subsetでcolumn名を指定してあげる

rt2 = df.dropna(subset=['runtime'])
rt2
```

### .fillna()

欠損値(nan)を()で指定した物で埋める

```python
# 今回は欠損値をruntimeの平均値で埋める
# 平均値を算出する

df['runtime'].mean()
```

```python
# fillnaで平均値で埋めてあげたdf['runtime']を
# df['runtime']に代入してあげることで、元データのdfが更新される

df['runtime'] = df['runtime'].fillna(df['runtime'].mean())
df['runtime']
```

```python
# nanが無くなったか確認

df[np.isnan(df['runtime'])]
```

### .isna()

dataframeでnanのものをTrueで返してくれる

```python
df = pd.read_csv('movie_dataset/tmdb_5000_movies.csv')
df
```

```python
df.isna()
```

```python
# columnsの指定も可能

df['homepage'].isna()
```

```python
# フィルターとしても使える
# homepageがnanの物のみを返してくれる
# df[np.isnan(df['homepage'])]と同じ

df[df['homepage'].isna()]
```

```python
# homepageがnanの物以外を返してくれる

df[~(df['homepage'].isna())]
```

```python
# 全体のデータ数と合うか検算

a = df[df['homepage'].isna()]# runtimeでnanのもの
b = df[~(df['homepage'].isna())]# runtimeでnan以外のもの

c = len(a) + len(b)
c# 4803
```

```python
# df['homepage'].isna()と同じ

pd.isna(df['homepage'])
```

```python
# 別の変数に代入して、新しいcolumnsとしてdataframeに足すことも可能

df['homepage_nan'] = df['homepage'].isna()
```
