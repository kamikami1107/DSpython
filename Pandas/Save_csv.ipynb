```python
import numpy as np
import pandas as pd
```

### .to_csv()

- profit(revenue - budget)を降順に並べた表を作成してみる

- budgetやrevenueが0のものは除く



```python
df = pd.read_csv('movie_dataset/tmdb_5000_movies.csv')
```

```python
df.head()
```

```python
# df,dropnaで削る為に0をnanに変える関数を作成

def return_nan_if_zero(a):
    """
    Parameters
    ----------
    budget:int
    予算

    Returns
    -------
    return_nan_if_zero(0):nan
    return_nan_if_zero(100000):100000
    """
    
    if a == 0:
        return np.nan
    else:
        return a
```

```python
return_nan_if_zero(0)
```

```python
return_nan_if_zero(100)
```

```python
# 上記をlambda関数で作成
return_nan_if_zero_lambda = lambda x: np.nan if x==0 else x
```

```python
return_nan_if_zero_lambda(0)
```

```python
return_nan_if_zero_lambda(100)
```

```python
# budgetにapply
df['budget'] = df['budget'].apply(return_nan_if_zero)
df['budget']
```

```python
# revenueにapply
df['revenue'] = df['revenue'].apply(return_nan_if_zero)
df['revenue']
```

```python
df = df.dropna(subset=['budget', 'revenue'])
df
```

```python
# profit(budget - revenue)を作成する前に要素数が同じものか確認

print(len(df['budget']))
print(len(df['revenue']))
```

```python
# 0をnanに変えてdropnaで落とすのは煩わしい
# フィルターアウトで一気に消す

# df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]
```

```python
# profitというcolumnsを作成

df['profit'] = df['revenue'] - df['budget']
```

```python
# 確認
df.head()
```

```python
# profitを降順にする

df.sort_values('profit', ascending=False).head()
```

```python
# csvファイルを作成
# index=Falseにすると元のindexが保持されずに保存できる
df.to_csv('movie_dataset/tmdb_5000_movies_profit_sorted.csv', index=False)
```
