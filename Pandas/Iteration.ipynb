### .apply()

````python
# revennue - budgetが各映画の収益
# 収益を求める
````

````python
# 各カラムに0があると正しく無い情報が計算されてしまうので、Nanに置き換える
# 前提として以下の関数を作成

def return_nan_if_zero(budget):
    """
    Parameters
    ----------
    budget:int
    予算

    Returns
    -------
    return_nan_if_zero(0):nan
    return_nan_if_zero(100000):100000
    """
    if budget == 0:
        return np.nan
    else:
        return budget
````

````python
return_nan_if_zero(0)
````

````python
return_nan_if_zero(100000)
````

````python
# 上記の関数を.apply()に代入してdfの各行に適用
# 今回はbudgetに適用

df['budget'].apply(return_nan_if_zero)
````

````python
# 元のデータを更新

df['budget'] = df['budget'].apply(return_nan_if_zero)
````

````python
# 確認
df.tail(3)
````

````python
# budgetの0がnanに更新されているので、再ダウンロード

df= pd.read_csv('movie_dataset/tmdb_5000_movies.csv')
````

````python
# budgetでフィルタリングした物を昇順にし、確認

df.sort_values('budget', ascending=False)
````

````python
# 作ったlambda関数を.apply()に代入

df['budget'] = df['budget'].apply(lambda a :np.nan if a == 0 else a)
df['revenue'] = df['revenue'].apply(lambda b :np.nan if b ==0 else b)
````

````python
# Nanに置き換わったか確認

df.sort_values('budget', ascending=False)
````

````python
# データがnanに置き換わったbudgetとrevenueをdrop
df = df.dropna(subset=['budget', 'revenue'])
````

````python
# 目視確認
df.tail()
````

````python
# 元のデータフレームよりも数が減っていれば、しっかりとdropされている
len(df)
````

````python
# 元の話題に戻る
# revenue - budgetが各映画の収益
# 収益を求める
````

````python
# profit = revenue - budget
df['profit'] = df['revenue'] - df['budget']
df['profit']
````
