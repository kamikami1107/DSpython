```python
import pandas as pd
import numpy as np
```

## `pd.DataFrame()`

### Numpy Arrayから作成

```python
ndarray1 = np.arange(10).reshape(2, 5)
print(ndarray1)
type(ndarray1)
```

```python
ndarray2 = np.linspace(0, 9, num=10).reshape(2, 5)
print(ndarray2)
type(ndarray2)
```

```python
pd.DataFrame(ndarray1)
```

```python
pd.DataFrame(ndarray2)
```

```python
pd.DataFrame(ndarray1, 
            index = ['index1', 'index2'], 
            columns = ['a', 'b', 'c', 'd', 'e'])
```

### Dictionaryから作成

```python
data1 = {
    'name': 'John',
    'sex': 'male',
    'age': 22
}
data2 = {
    'name': 'Zack',
    'sex': 'male',
    'age': 30
}
data3 = {
    'name': 'Emily',
    'sex': 'female',
    'age': 32
}
pd.DataFrame([data1, data2, data3])
```

```python
# more easier

data = {
    'name': ['John', 'Zack', 'Emily'], 
    'sex': ['male', 'male', 'female'], 
    'age': [22, 30, 32]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# 稀なケース
# indexがStringsで、columnsが数字の場合

columns = [1, 2, 3, 4, 5]
index = ['a', 'b', 'c', 'd', 'e']

array = np.arange(25).reshape(5, 5)

df = pd.DataFrame(array, index=index, columns=columns)
```

```python
data = {
    'name': ['John', 'Zack', 'Emily'], 
    'sex': ['male', 'male', 'female'], 
    'age': [22, 30, 32]
}
df = pd.DataFrame(data)
```

```python
# 新しいcolumnsを追加する

df['cost'] = [100, 200, 300]
```

```python
# 特定のindexやcolumnsを削る

df.drop(0)# index除去
```

```python
df.drop('name', axis=1)# columns除去
```

```python
# inplace=Trueでオリジナルデータの該当データも除去

df.drop('name', axis=1, inplace=True)
```
