## Index操作

### .reset_index()で再度indexを割り振る

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_dataset/tmdb_5000_movies.csv')
df.head()
```

```python
df[df['original_language'] == 'ja'].head()
```

```python
# indexがリセットされる

df.reset_index()
```

```python
# indexをリセットして、更にindex自体も消す
# ただしindexをIDのように使用しているケースもあるので、
# その場合は誤って消さないように注意する

df.reset_index(drop=True)
```
