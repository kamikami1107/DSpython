### 保存と読み込み

```python
import numpy as np
```

```python
np.random.seed(1)
ndarray = np.random.randn(3, 4, 5)
ndarray
```

```python
np.save('sample_ndarray', ndarray)# 自動的に.npyの拡張子がついて保存される
```

```python
np.load('sample_ndarray.npy')# loadのときは.npyを忘れずに
```

### 辞書を用いた場合

```python
dictionary = {'id': 123456, 
              'image': ndarray}
```

```python
dictionary
```

```python
np.save('sample_dict', dictionary)
```

```python
loaded_dict = np.load('sample_dict.npy', allow_pickle=True)
```

```python
loaded_dict
```

```python
a = np.array({'a':1})
```
